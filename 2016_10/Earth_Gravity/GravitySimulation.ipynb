```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.pyplot import cm
%matplotlib inline
```

```python
dat = pd.read_csv("output.csv")
```

```python
N=100 # 一辺の長さ
```

```python
X = np.array(dat['X'])
Y = np.array(dat['Y'])
U = np.array(dat['U'])
V = np.array(dat['V'])

plot3 = plt.figure()
plt.quiver(X, Y, -U, -V)
plt.savefig("vector.png",format = 'png', dpi=1000)
```

```python
X = np.array(dat['X']).reshape(N,N)
Y = np.array(dat['Y']).reshape(N,N)
U = np.array(dat['U']).reshape(N,N)
V = np.array(dat['V']).reshape(N,N)

plot3 = plt.figure()
Z =  np.sqrt(U*U+ V*V)
plt.contourf(X, Y,Z)
plt.savefig("contourf.png",format = 'png')
```

```python
X = np.array(dat['X']).reshape(N,N)
Y = np.array(dat['Y']).reshape(N,N)
U = np.array(dat['U']).reshape(N,N)
V = np.array(dat['V']).reshape(N,N)
plt.plot(Y[N/2,:],V[N/2,:])
plt.savefig("linear.png")
```

- 上記3つのグラフから次のことが読み取れる
    + 力のベクトルは中心から真っすぐ外側に伸びている(当たり前)
    + 重力が最も大きくなるのは地表である
    + 地球内部については線形的な復元力となっている(地球外部では非線形，多分距離の二乗に反比例)
    + 地球にほったトンネルを走る電車はちゃんと単振動する
