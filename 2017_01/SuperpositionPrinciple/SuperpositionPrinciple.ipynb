# 数値計算で重ね合わせの原理の検証(お遊び)

周期の異なる2つの調和外力によるバネマスダンパー系の強制振動を表した運動方程式の定常応答を考える．
$$
m\frac{d^2 x}{dt^2} + c\frac{d x}{dt} + kx = f_1 \cos(w_1 t + \phi_1) + f_2 \cos(w_2 t + \phi_2)
$$
この運動方程式の定常応答は，重ね合わせの原理より以下の2つの運動方程式の解 $x_1,x_2$ の和として表されることが知られている．
$$
m\frac{d^2 x_1}{dt^2} + c\frac{d x_1}{dt} + kx_1 = f_1 \cos(w_1 t + \phi_1) 
$$
$$
m\frac{d^2 x_2}{dt^2} + c\frac{d x_2}{dt} + kx_2 = f_2 \cos(w_2 t + \phi_2)
$$
これが正しいかどうか，数値計算によって確認する．

```python
# 前準備
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

%matplotlib inline
```

```python
# 各パラメータを設定する．
m = 20
c = 30
k = 2000
f_1 = 20
f_2 = 20
w_1 = np.sqrt(k/m)*2  # 固有角振動数の2倍
w_2 = np.sqrt(k/m)*3  # 固有角振動数の3倍
phi_1 = np.pi / 2
phi_2 = np.pi / 4
t = np.linspace(0,10,num=1e3)
```

先の運動方程式を連立微分方程式の形に書き下す．右辺をF(t)と置くと次のように変形できる．
$$
\frac{d}{dt}
\left[
    \begin{array}{rrr}
     x \\
     \dot x
    \end{array}
 \right] = 
 \left[
    \begin{array}{rrr}
     \dot x \\
     \frac{-c \dot x - k x + F(t)}{m}
    \end{array}
 \right]
$$

```python
# ベクトル[ x  (dx/dt)]を受け取ってその微分を返す関数
def fprime(x,t,F):
    dxdt = np.zeros(2)
    dxdt[0] = x[1]
    dxdt[1] = (-c*x[1]-k*x[0]+F(t))/m
    return dxdt
```

外力なし，調和外力1，調和外力2，調和外力1と2を表す関数を定義する．

```python
F0 = lambda t: 0
F1 = lambda t: np.cos(w_1*t+phi_1)
F2 = lambda t: np.cos(w_2*t+phi_2)
F3 = lambda t: F1(t)+F2(t)
```

試しに減衰ありの初期変位あり外力なしの場合における減衰振動を見てみると次のようになる．
振幅が指数関数的に減っていくのがわかる．

```python
x_0 = odeint(fprime,[0.2,0.0],t,args=(F0,)) # 初期変位0.2，初期速度0.0として計算
plt.plot(t,x_0[:,0],label="no force")
```

次に外力1がある状態での挙動を見てみる．

```python
x_1 = odeint(fprime,[0.0,0.0],t,args=(F1,))
plt.plot(t,x_1[:,0],label="f1 force")
```

次に外力2がある状態での挙動を見てみる．初期変位，初期速度は0とした．

```python
x_2 = odeint(fprime,[0.0,0.0],t,args=(F2,))
plt.plot(t,x_2[:,0],label="f2 force")
```

外力1と2がある状態での挙動を見る．初期変位，初期速度は0．同じグラフにf1の解，f2の解を足し合わせたものを表示させてみたところ，概ね一致していることが読み取れる．

```python
x_3 = odeint(fprime,[0.0,0.0],t,args=(F3,))
plt.plot(t,x_3[:,0],label="f1+f2 force")
plt.plot(t,x_1[:,0]+x_2[:,0],label="f1 sol.+f2 sol")
plt.legend()
```

過渡応答がある場合として，初期変位0.0002，初速度0.0として計算してみる．

```python
x_init = [2e-4,0]
x_0 = odeint(fprime,x_init,t,args=(F0,))
x_1 = odeint(fprime,x_init,t,args=(F1,))
x_2 = odeint(fprime,x_init,t,args=(F2,))
x_3 = odeint(fprime,x_init,t,args=(F3,))
```

外力1と2がある状態での挙動と，f1の解，f2の解を足し合わせたものを表示させてみる．過渡応答x_0がx_1,x_2に含まれているのではじめは一致しないが，時間が経つとx_0が0に収束するため一致する．x_1+x_2-x_0とすればx_3と一致する．

```python
plt.plot(t,x_3[:,0],label="f1+f2 force")
plt.plot(t,x_1[:,0]+x_2[:,0],label="f1 sol.+f2 sol")
plt.legend()
```
